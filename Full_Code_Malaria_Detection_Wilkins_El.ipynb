```python
#Storing the path for train
train_path = '/content/cell_images/train'
# resizing images
SIZE = 64
train_images = []
train_labels = [] #0 - uninfected, 1 - parasitized


#Storing the path for test
test_path = '/content/cell_images/test'
test_images = []
test_labels = []
```

```python
# Method that will help with training test and train images
def load_images_and_labels(data_path, size):
    images = []
    labels = []

    for folder_name in ['/parasitized/', '/uninfected/']:
        images_path = os.listdir(data_path + folder_name)

        for i, image_name in enumerate(images_path):
            try:
                # image = Image.open(data_path + folder_name + image_name)
                image = cv2.imread(data_path + folder_name + image_name)
                # image = image.resize((size, size))
                image = cv2.resize(image, (size, size))
                images.append(np.array(image))

                if folder_name == '/parasitized/':
                    labels.append(1)
                else:
                    labels.append(0)
            except Exception:
                pass

    return np.array(images), np.array(labels)
```

```python
# Load test data
# Load training data
train_images, train_labels = load_images_and_labels(train_path, SIZE)
test_images, test_labels = load_images_and_labels(test_path, SIZE)
```

###<b> Importing the required libraries for building and training our Model

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dropout, Dense, Flatten
from keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix
import numpy as np
import shutil
import os
import glob
```

```python
image_paths = glob.glob("/content/cell_images/train/*/*.png")
labels = list(map(lambda x: x.split('/')[-2], image_paths))
```

```python
# The split is performed with a test/validation size of 20% of the data, and it uses stratification to maintain the class distribution in the split.
train_image_paths, val_image_paths = train_test_split(image_paths, test_size=0.2, stratify=labels)
```

```python
TRAIN_IMAGE_FOLDER = "training_data"
VALIDATION_IMAGE_FOLDER = "validation_data"
os.makedirs(f"/content/cell_images/{TRAIN_IMAGE_FOLDER}/parasitized", exist_ok=True) #making directories
os.makedirs(f"/content/cell_images/{TRAIN_IMAGE_FOLDER}/uninfected", exist_ok=True)

os.makedirs(f"/content/cell_images/{VALIDATION_IMAGE_FOLDER}/parasitized", exist_ok=True)
os.makedirs(f"/content/cell_images/{VALIDATION_IMAGE_FOLDER}/uninfected", exist_ok=True)

for img_path in train_image_paths:
  shutil.move(img_path, img_path.replace("train", "training_data")) #renaming

for img_path in val_image_paths:
  shutil.move(img_path, img_path.replace("train", "validation_data"))
```

```python
# data is only rescaled to have pixel values between 0 and 1
datagen = ImageDataGenerator(rescale=1./255,
                                      # zoom_range=0.2,
                                      # horizontal_flip=True,
                                      # vertical_flip=True,
                                      # width_shift_range=0.2,
                                      # height_shift_range=0.2,
                                      # validation_split=0.2
                                      )
train_data = datagen.flow_from_directory(
    f"/content/cell_images/{TRAIN_IMAGE_FOLDER}",
    target_size=(64,64),
    batch_size=128,
    class_mode = 'binary',
)

validation_data = datagen.flow_from_directory(
    f"/content/cell_images/{VALIDATION_IMAGE_FOLDER}",
    target_size=(64,64),
    batch_size=128,
    class_mode = 'binary',
    shuffle=False
)

test_data = datagen.flow_from_directory(
    "/content/cell_images/test",
    target_size=(64,64),
    batch_size=128,
    class_mode = 'binary',
    shuffle=False
)
```

###<b> Building the model

```python
model=Sequential([
    Conv2D(16,(3,3),input_shape=(64,64,3),activation='relu'),
    MaxPool2D(2,2),

    Conv2D(32,(3,3),activation='relu'),
    MaxPool2D(2,2),

    # Conv2D(64,(3,3),activation='relu'),
    # MaxPool2D(2,2),


    Flatten(),
    # Dense(256,activation='relu'),
    Dense(1,activation='sigmoid')

])
```

###<b> Compiling the model

```python
model.compile(loss='binary_crossentropy',
              optimizer='Adam',
              metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=5, verbose=1, restore_best_weights=True)
```

<b> Fit and train our Model

```python
history = model.fit_generator(
    train_data,
    validation_data = validation_data,
    epochs = 20,
    verbose = 1,
    callbacks=[early_stopping]
)
```

###<b> Evaluating the model on test data

```python
evaluation = model.evaluate_generator(test_data)

loss = evaluation[0]
accuracy = evaluation[1]
print(f'Loss: {loss}, Accuracy: {accuracy}')
```

<b> Callbacks

```python
# Generate predictions using the model and the data generator
predictions = model.predict_generator(test_data)

# Convert predicted probabilities to binary class labels (0 or 1)
predicted_labels = np.round(predictions)

# Obtain the true binary labels from the data generator
true_labels = test_data.classes

# Calculate the confusion matrix
confusion = confusion_matrix(true_labels, predicted_labels)
print(confusion)
```

<b>Plotting the train and validation curves

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()


acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
